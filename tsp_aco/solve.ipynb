```python
import numpy as np
import elkai
from scipy.spatial import distance_matrix

SCALE = 1000000


path = f"datasets/test_TSP100.npy"
dataset = np.load(path)
n_instances = dataset.shape[0]
optimal = []
for i, instance in enumerate(dataset):
    dist = distance_matrix(instance, instance)
    elkai_dist = elkai.DistanceMatrix((dist * SCALE).astype(int).tolist())
    opt_route = elkai_dist.solve_tsp(runs=20)
    opt_cost = 0
    for i in range(len(opt_route) - 1):
        opt_cost += dist[opt_route[i], opt_route[i + 1]]
    opt_cost += dist[opt_route[-1], opt_route[0]]
    optimal.append(opt_cost)
mean_optimal = np.mean(optimal)
total = np.sum(optimal)
print(f"Total optimal cost for TSP50: {total}")
print(f"Mean optimal cost for TSP50: {mean_optimal}")
```

```python
import numpy as np
import elkai
from scipy.spatial import distance_matrix

SCALE = 1000000

for size in [20, 50, 100, 200]:
	path = f"datasets/test_TSP{size}.npy"
	dataset = np.load(path)
	n_instances = dataset.shape[0]
	optimal = []
	for i, instance in enumerate(dataset):
		dist = distance_matrix(instance, instance)
		elkai_dist = elkai.DistanceMatrix((dist * SCALE).astype(int).tolist())
		opt_route = elkai_dist.solve_tsp(runs=20)
		opt_cost = 0
		for i in range(len(opt_route) - 1):
			opt_cost += dist[opt_route[i], opt_route[i + 1]]
		opt_cost += dist[opt_route[-1], opt_route[0]]
		optimal.append(opt_cost)
	mean_optimal = np.mean(optimal)
	total = np.sum(optimal)
	print(f"Mean optimal cost for TSP{size}: {mean_optimal}")
```
